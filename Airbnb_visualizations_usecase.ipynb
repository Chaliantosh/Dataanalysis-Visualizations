```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Visualize Qualitative data

```python
listings=pd.read_csv(r'C:\Users\admin\Desktop\Airbnb_NYC_2019.csv')
listings
```

# Countplot

```python
#Barchart that displays count of Airbnb in each neighbourhood group.
sns.countplot(x='neighbourhood_group',data=listings)
plt.show()
```

Observations: Manhattan neighbourhood group is the highest count of Airbnb listings in NYC in 2019. Follwed by Brooklyn, Queens, Bronx and Staten Island.

# Barplot

```python
# Average price in each neighbourhood group in NYC
sns.barplot(x='neighbourhood_group', y='price',data=listings)
plt.show()
```

```python
#recreate barplot without black lines
sns.barplot(x='neighbourhood_group', y='price',data=listings, ci=False)
plt.show()
```

Observations: Manahattan neighbourhood group in NYC has the highest average price.

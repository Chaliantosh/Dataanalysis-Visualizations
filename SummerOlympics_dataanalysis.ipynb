```python
#Q16. Plot the number of gold medals won by the US male & female athletes throughout the history of the Olympics?
gog = oo[(oo.Medal == 'Gold') & (oo.NOC == 'USA')].groupby(['Edition', 'Gender']).size() #.plot(kind='bar');
gog
```

```python
#Q17. Plot the five athletes who have won the most gold medals over the history of the Olympics. When there is a tie, 
#     consider the number of silver medals, and then bronze medals.
g = oo.groupby(['Athlete', 'Medal']).size().unstack('Medal',fill_value=0)
h = g.sort_values(['Gold', 'Silver', 'Bronze'],ascending=False)
i = h[['Gold','Silver','Bronze']]
i.head().plot(kind='bar');
```

```python
#Q18: From Q17, change the color of gold, Silver and Bronze to provide more intutive meaning to the graph.
from matplotlib.colors import ListedColormap
gsb=['#dbb40c', '#c5c9c7', '#a87900']
my_gsb = ListedColormap(sns.color_palette(gsb))
```

```python
i.head().plot(kind='bar', colormap=my_gsb);
```

```python
#Q19: In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.
gy = oo[oo.NOC == 'USA'] #.groupby(['Athlete','Medals']).size()
gy_med = gy.groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value=0)
gy_med['Total'] = gy_med['Gold'] + gy_med['Bronze'] + gy_med['Silver']
gy_med.reset_index(inplace=True)
tu = [group.sort_values('Total',ascending=False)[:1] for year, group in gy_med.groupby('Edition')]
tu
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```

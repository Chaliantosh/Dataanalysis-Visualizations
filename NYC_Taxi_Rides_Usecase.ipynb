# Working with Categorical Data

```python
df.VendorID.unique()
```

```python
#create a mapping for these values using a dictionary
df['Vendor']=df['VendorID'].apply({1:'Creative', 2:'Verifone'}.get)
```

```python
df
```

```python
#check the memory consumed by this mapping
df['Vendor'].memory_usage() / (1<<20)
```

```python
#create a mapping for these values using a dictionary with type as category
df['Vendor']=df['VendorID'].apply({1:'Creative', 2:'Verifone'}.get).astype('category')
```

```python
df
```

```python
#check the memory consumed by this mapping after converting to categorical type
df['Vendor'].memory_usage() / (1<<20)
```

Lot of memory conserved by switching to categorical data

```python
#underlying data can be accessed thru .code attribute
df['Vendor'].cat.codes
```

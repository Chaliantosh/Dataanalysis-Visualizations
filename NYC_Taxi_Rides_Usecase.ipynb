# Examine Data

```python
df.iloc[0]
```

Obervation: Data doesn't look correct, For e.g. Vendor ID attribute looks like a time, Store_and_fwd_flag looks like a coordinate etc.

```python
with bz2.open(fname,'rt') as fp:
    header=fp.readline()
    data=fp.readline()
    
print(header)
print(data)
```

Observation: There are two commas (,,) at the end in the data.

```python
len(header.split(','))
```

```python
len(data.split(','))
```

Observation: Data has two extra fields, to fix this let us load only first 21 columns of data.

```python
import numpy as np
df=pd.read_csv(fname,usecols=np.arange(21))
df
```

```python
df.iloc[0]
```

Observations: Now the Vendor ID and other columns look better.

```python
df.dtypes
```

Observations: Attributes, 'lpep_pickup_datetime' & 'Lpep_dropoff_datetime' are timestamps but mentioned as object.

```python
df=pd.read_csv(fname,usecols=np.arange(21),parse_dates=(['lpep_pickup_datetime','Lpep_dropoff_datetime']))
df
```

```python
df.dtypes
```

Observations: Attributes, 'lpep_pickup_datetime' & 'Lpep_dropoff_datetime' are now datetime.

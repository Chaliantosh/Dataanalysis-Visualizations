```python
import numpy as np
from numpy.random import randn
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline 
rcParams['figure.figsize']=10,4
```

```python
cars=pd.read_csv(r'C:\Users\admin\Desktop\mtcars.csv')
cars.rename(columns={'Unnamed: 0':'car_names'},inplace=True)
cars
```

```python
#plotting a line chart with one variable 'mpg' 
mpg = cars.mpg
mpg.plot();
```

```python
#plotting a line xhart with three variables 'cyl', 'wt', 'mpg'
df = cars[['cyl','wt','mpg']]
df.plot();
```

```python
#plotting a bar chart with one variable 'mpg' 
cars.mpg.plot(kind='bar');
```

```python
#plotting a horizantal bar chart with one variable 'mpg' 
cars.mpg.plot(kind='barh');
```

```python
#adding user defined colors to the plot
color_theme = ['darkgray','lightsalmon','powderblue']
df.plot(color=color_theme);
```

```python
#object oriented plotting of cars_names in x axes vs mpg

fig = plt.figure() # creating a blank figure
ax = fig.add_axes([.1,.1,1,1]) # axes to figure

mpg.plot();

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Fallon of Cars in mtcars Dataset')

ax.set_xlabel('Car names')
ax.set_ylabel('miles/gal')
```

```python
#add a legened to the plot.
fig = plt.figure() # creating a blank figure
ax = fig.add_axes([.1,.1,1,1]) # axes to figure

mpg.plot();

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Fallon of Cars in mtcars Dataset')

ax.set_xlabel('Car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
```

```python
#adding annotations to the plot
#add a legened to the plot.
fig = plt.figure() # creating a blank figure
ax = fig.add_axes([.1,.1,1,1]) # axes to figure

mpg.plot();

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Fallon of Cars in mtcars Dataset')

ax.set_xlabel('Car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')

ax.set_ylim([0,45])

ax.annotate('Toyota Corolla', xy=(19,33.9),xytext=(21,35),
          arrowprops=dict(facecolor='black',shrink=0.05))
```

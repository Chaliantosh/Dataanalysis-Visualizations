# Q1: what were the 5 most popular baby names in 2014 in the US?
Approach:
    1. Slice out rows for 2014 using boolean indexing
    2. Sort rows in descending order by count.
    3. Get the first five rows.            

```python
#1. slice out rows for 2014.
us_b2014 = us_babies[us_babies.Year==2014]
us_b2014
```

# Q3: In CA, what were the most popular female & male baby names in each yr?
Approach:
    1. Slice out rows for CA state using boolean indexing
    2. Group the resulting DataFrame by 'Year' & 'Gender'
    3. Compute the most popular name for each group.    

```python
ca_babies=states_babies[states_babies.State=='CA']
ca_babies
```

```python
def popular(s):
    """ Receies s, a Pandas Series, containing baby names in order of highest count to lowest count. Returns the most popular
    name in s."""
    return s.iloc[0]
```

```python
ca_grouped=ca_babies.sort_values('Count',ascending=False).groupby(['Year','Gender']).agg(popular)
ca_grouped
```
